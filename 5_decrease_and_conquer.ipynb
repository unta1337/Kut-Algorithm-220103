```python
def main():
    # 삽입정렬의 구현.
    print('# 삽입정렬의 구현\n')

    n = 10
    lst = [random.randint(1, n) for _ in range(n)]

    print(f'Initial List: {lst}\n')

    print('Bottom-Up:')
    insertion_sort_bottom_up(lst.copy())
    print()

    print('Top-Down:')
    insertion_sort_top_down(lst.copy())
    print()

    # 위상정렬의 구현.
    print('# 위상정렬의 구현\n')

    print('# 축소정복 기법')
    print('본문 그래프: ', end = '')
    for vertex in topological_sort_decrease_and_conquer(graph_in_content):
        print(vertex, end = ' ')
    print()

    print('문제 그래프 1: ', end = '')
    for vertex in topological_sort_decrease_and_conquer(graph_in_problem_1):
        print(vertex, end = ' ')
    print()

    print('문제 그래프 2: ', end = '')
    for vertex in topological_sort_decrease_and_conquer(graph_in_problem_2):
        print(vertex, end = ' ')
    print('\n')

    print('# DFS 기법: ')
    print('본문 그래프: ', end = '')
    for vertex in topological_sort_depth_first_search(graph_in_content):
        print(vertex, end = ' ')
    print()

    print('문제 그래프 1: ', end = '')
    for vertex in topological_sort_depth_first_search(graph_in_problem_1):
        print(vertex, end = ' ')
    print()

    print('문제 그래프 2: ', end = '')
    for vertex in topological_sort_depth_first_search(graph_in_problem_2):
        print(vertex, end = ' ')
```

```python
import random

graph_in_content = {
    'A': {'C', 'D'},
    'B': {'D', 'E'},
    'C': {'D', 'F'},
    'D': {'F'},
    'E': {'F'},
    'F': {}
}

graph_in_problem_1 = {
    'A': {'B', 'D', 'E'},
    'B': {'C', 'E', 'F'},
    'C': {'G'},
    'D': {'E'},
    'E': {},
    'F': {'G'},
    'G': {}
}

graph_in_problem_2 = {
    'A': {'B', 'C'},
    'B': {'D'},
    'C': {'D', 'E'},
    'D': {'F'},
    'E': {'G', 'H'},
    'F': {},
    'G': {'H'},
    'H': {'C'}
}
```

```python
# 1. 삽입정렬의 구현

# 1.1. 상향식 구현 (교재 4장 p. 177)
def insertion_sort_bottom_up(lst):
    for i in range(1, len(lst)):
        key = lst[i]
        j = i - 1
        while j >= 0 and lst[j] > key:
            lst[j + 1] = lst[j]
            j -= 1
        lst[j + 1] = key

        print(f'Current List: {lst}')
```

```python
# 1.2 하향식 구현
def insertion_sort_top_down(lst):
    __insertion_sort_top_down(lst, len(lst))

def __insertion_sort_top_down(lst, n):
    if n == 1:
        return

    __insertion_sort_top_down(lst, n - 1)

    key = lst[n - 1]
    index = n - 2
    while index >= 0 and lst[index] > key:
        lst[index + 1] = lst[index]
        index -= 1
    lst[index + 1] = key

    print(f'Current List: {lst}')
```

##### # 상향식 삽입정렬과 하향식 삽입정렬의 비교분석  

\# 삽입정렬  

삽입정렬은 첫 두 원소부터 시작하여 뒤에 오는 원소의 위치를 결정하여 정렬하는 정렬 알고리즘이다.  

예를 들어,  
$$[5, 4, 3, 2, 1]$$
위와 같은 리스트가 있다고 하자.  
삽입정렬을 시행하면 첫 두 원소로 5와 4가 선택된다. 이때 4의 위치를 결정하는데, 앞의 원소와 비교하여 4보다 작거나 같은 원소가 나오면 해당 원소를 한 칸 뒤로 밀어 4가 들어갈 공간을 마련한다.  
만약 조건이 만족하지 않아 인덱스가 0이 되면 0에 4를 넣으면 된다.  

위의 예시에서는 5가 4보다 크므로 인덱스가 0인 상태에서 조건이 만족하지 않으므로 4를 0번 인덱스에 넣으면 된다.  
$$[4, 5, 3, 2, 1]$$
가 된다.  

같은 방법으로 계속해서 진행하면,  
$$[3, 4, 5, 2, 1]$$
$$[2, 2, 4, 5, 1]$$
$$[1, 2, 3, 4, 5]$$
와 같이 리스트를 정렬할 수 있다.  

삽입정렬은 리스트의 길이를 n이라 했을 때, 각 단계에서 대략 n개의 원소를 비교한다.  
더불어 해당 단계를 약 n번 반복하므로 복잡도는,  
$$O(N^2)$$
이 된다.  

\# 상향식과 하향식  

분할정복 정렬인 병합정렬의 경우, 구간을 나누는 방법에 따라서 상향식과 하향식에서 나뉘는 구간의 길이 등이 변할 수 있다.  
그러나 삽입정렬은 상향식과 하향식이 동일한 동작을 보인다.  

상향식의 경우 부분 리스트의 길이가 2부터 시작하여 n까지 가는 반면, 상향식은 n에서 시작하여 1로 간다.  
그런데 실제로 알고리즘이 수행되는 순서는 리스트의 길이가 1, 2, ..., n의 순서로 동작하므로 상향식과 하향식의 동작 방식은 재귀호출을 제외하면 완전히 동일하다.

```python
# 2. 위상정렬의 구현

# 2.1 축소정복 기법
def topological_sort_decrease_and_conquer(graph):
    # 진입차수가 0인 정점을 먼저 탐색하기 위해 해당되는 정점 리스트 생성.
    in_degrees = {v: 0 for v in graph}
    for v in graph:
        for u in graph[v]:
            in_degrees[u] += 1
    zero_degree_vertices = [v for v in graph if in_degrees[v] == 0]
    path = []

    # 진입차수가 0인 정점에 대해 수행.
    while zero_degree_vertices:
        current_vertex = zero_degree_vertices.pop()
        path.append(current_vertex)        # 현재 정점 삽입. (위상정렬의 결과)

        # 현재 정점에서 나가갈 수 있는 다음 정점에 대해 수행.
        for next_vertex in graph[current_vertex]:
            # 현재 정점은 제거될 것이므로 다음 정점에 대해 진입차수 1 감소.
            in_degrees[next_vertex] -= 1

            # 다음 정점의 진입차수가 0이 되면 진입차수가 0인 리스트에 추가.
            if in_degrees[next_vertex] == 0:
                zero_degree_vertices.append(next_vertex)

    # 경로의 길이가 그래프 정점의 개수보다 작으면 위상정렬이 이뤄지지 않은 것임.
    return path if len(path) == len(graph) else [None]
```

```python
# 2.2 DFS 기법
def topological_sort_depth_first_search(graph):
    # 진입차수가 0인 정점을 먼저 탐색하기 위해 해당되는 정점 리스트 생성.
    in_degrees = {v: 0 for v in graph}
    for v in graph:
        for u in graph[v]:
            in_degrees[u] += 1
    zero_degree_vertices = [v for v in graph if in_degrees[v] == 0]
    visited = {v: False for v in graph}
    path = []

    # 진입차수가 0인 정점에 대해 DFS 수행.
    for v in zero_degree_vertices:
        __topological_sort_decrease_and_conquer(graph, in_degrees, v, visited, path)
    
    # 경로의 길이가 그래프 정점의 개수보다 작으면 위상정렬이 이뤄지지 않은 것임.
    return reversed(path) if len(path) == len(graph) else [None]    # 들어가는 순서가 아닌 나오는 순서대로가 역-위상순서이므로 이를 반전하여 반환.

def __topological_sort_decrease_and_conquer(graph, in_degrees, current_vertex, visited, path):
    # 이미 살펴본 정점이면 백트랙킹.
    if visited[current_vertex]:
        return
    visited[current_vertex] = True

    # 정점의 진입차수 갱신.
    in_degrees[current_vertex] -= 1
    if in_degrees[current_vertex] > 0:
        path.append(current_vertex)     # 더 이상 진행할 수 없으므로 하나의 종점으로 취급, 해당 정점을 경로에 포함한다.
        return

    # 다음 정점에 대해 재귀적으로 DFS 시행.
    for next_vertex in graph[current_vertex]:
        __topological_sort_decrease_and_conquer(graph, in_degrees, next_vertex, visited, path)
    # 참고: 이미 방문한 정점을 다시 방문하지 않으므로 진입차수를 원상태로 돌려놓을 필요가 없다.

    # 전위 순회 시에는 백트랙킹에 의해 경로가 꼬일 수 있다.
    # 즉, 위상정렬의 최종 경로에서 정점 간의 길이 연결됨을 보장할 수 없다.
    # 후위 순회는 말단 노드부터 루트 노드까지 한 번에 연결되므로 정점 간의 길이 연결됨을 보장한다.
    path.append(current_vertex)
```

##### # 위상정렬의 축소정복 기법과 DFS로의 구현

\# 위상정렬 판별

위상정렬을 판별할 수 있는 요소에는 세 가지가 있다.  

첫째, 그래프의 부분 그래프에 순환 그래프가 있는 경우 해당 순환 그래프의 순서가 모호해져 위상정렬이 불가하다.  
둘째, 그래프를 일렬로 정렬하려면 진입차수가 0인 정점과 진출차수가 0인 정점이 적어도 한 개씩 존재해야 한다.  
셋째, 그래프를 일렬로 늘어놓았을 때 정점에서 뻗어나가는 경로가 항상 오른쪽을 향해야 한다. (왼쪽-오른쪽 순서로 정렬 시)  

위의 요소 외에도 위상정렬 가능 여부를 판단할 수 있는 방법이 하나 더 있다.  
바로 정렬된 그래프의 길이를 보는 것이다. 만약 위상정렬이 정상적으로 진행되어 정렬된다면, 정렬된 배열의 크기는 정점의 개수와 동일할 것이다.  
그러나 모종의 이유로 중간에 정렬을 중단한다면 정렬된 배열의 크기는 정점의 개수보다 작을 것이다.  

\# 축소정복 기법과 DFS 기법의 비교분석  

먼저 축소정복 기법과 DFS 기법은 접근 방법이 다르므로 두 결과가 다르게 나올 수 있다.
위상정렬의 경우의 수는 한 가지가 아니다. 더불어 어떤 정점을 먼저 탐색하느냐에 따라서 결과가 달라질 수 있다.  

축소정복 기법의 경우 항상 진입차수가 0일 때에만 경로에 포함하게 하므로 별다른 조치없이 경로를 계산할 수 있다.  
반면에 DFS 기법은 전위, 중위, 후위 순회에 따라서 결과가 달라지기 때문에 이를 고려해야 한다.  

전위와 중위 순회는 출력한 경로의 순서가 최종 순서가 될 것이라고 보장하지 못한다.  
전위의 경우 백트래킹 과정에서 경로가 변경될 수 있다.  
중위의 경우 그래프 순회는 트리 순회가 아니므로 순서에 대한 기준이 모호하므로 사용하기 어렵다.  

반면에 후위 순위의 경우 전위처럼 출력 이후에 추가적인 정점 탐색에 의한 순서 반전이 일어나지 않는다.  
이는 후위 순위의 경우 말단 노드부터 루트 노드까지 한 번에 진행하기 때문이다.  
이때 후위 순위의 순서대로 출력한다면 말단-루트의 경로가 되므로 루트-말단의 경로가 되도록 이를 반전해야 한다.
