````python
import random
from functools import reduce

# quick_selection 관련 변수.
lst = [i for i in range(5)] * 2
lst.sort(key = lambda e: random.random() < 0.5)

# 동전 문제 관련 변수 및 함수.
coins = [{'weight': 2, 'mark': None} for _ in range(11)]
coins.append({'weight': 1, 'mark': None})
# coins.append({'weight': 1, 'mark': None})
coins.sort(key = lambda e: random.random() < 0.5)

def scale(coins, indecies1, indecies2):
    w1 = reduce(lambda acc, e: acc + e['weight'], [coins[i] for i in indecies1], 0)
    w2 = reduce(lambda acc, e: acc + e['weight'], [coins[i] for i in indecies2], 0)

    return (w1 > w2) - (w1 < w2)

def mark(coins, indecies, mark):
    for i in indecies:
        coins[i]['mark'] = mark
````

````python
# 재귀적인 quick_select 함수의 원형.
def __quick_select_rec(lst, k, partition):
    return __quick_select_rec_core(lst, 0, len(lst) - 1, k, partition)

def __quick_select_rec_core(lst, left, right, k, partition):
    pos = partition(lst, left, right)       # partition 함수를 이용해 피봇 인덱스 추출.

    # 추출된 피봇이 원하는 순서의 원소이면 반환.
    if pos + 1 == left + k:
        return lst[pos]

    # 추출된 피봇이 원하는 순서의 원소보다 큰 경우 왼쪽 부분 리스트에 대해 quick_select 수행.
    # 이때 k번째 원소를 왼쪽에서부터 세었으므로 k를 바꿀 필요가 없음.
    elif pos + 1 > left + k:
        return __quick_select_rec_core(lst, left, pos - 1, k, partition)

    # 추출된 피봇이 원하는 순서의 원소보다 작은 경우 오른쪽 부분 리스트에 대해 quick_select 수행.
    # 이때 k번째 원소를 세는 기준이 변하게 되므로 k를 함께 바꾸어야 함.
    else:
        return __quick_select_rec_core(lst, pos + 1, right, k - (pos + 1 - left), partition)
````

````python
# 로무토의 분할 방법을 이용한 파티션 함수.
def __partition_lomuto(lst, left, right):
    pivot_value = lst[right]    # 기준값으로 리스트의 가장 오른쪽에 있는 값을 선택.
    l = left - 1                # r과 pivot_value의 관계에 의해 인덱스가 증가됨, 항상 pivot_value보다 작거나 같은 값을 가리킴.
    r = left                    # l의 증가에 관여함, 항상 pivot_value보다 큰 값을 가리킴.
    # Right Pivot, LL Pointers라고도 함.

    # r이 리스트의 끝에 다다를 때까지 반복.
    while r <= right:
        # 만약 r이 가리키는 원소가 l이 가리켜야 하는 원소일 경우.
        # 참고: r은 pivot_value보다 값이 클 때는 이를 무시하므로 r이 지나간 자리에는 pivot_value보다 큰 값만 존재함.
        #       단, r이 지나간 자리 중, l보다 작은 인덱스에는 pivot_value보다 작거나 같은 값만 존재함.
        if lst[r] <= pivot_value:
            l += 1      # l이 가리키는 원소는 항상 pivot_value보다 작음. 따라서 l + 1을 해주어 r의 값과 교환해도 문제가 없도록 함.

            lst[l], lst[r] = lst[r], lst[l]     # l과 r의 리스트값 교환.
        r += 1      # r은 조건에 부합하든 부합하지 않든 1씩 증가.

    # 마지막으로 l과 교환한 인덱스는 right이므로 lst[l] == pivot_value.
    return l
````

````python
# 호어의 분할 방법을 이용한 파티션 함수.
def __partition_hoare(lst, left, right):
    # 로무토의 방법과 동일하나 피봇을 왼쪽에서 뽑는다는 점과 l과 r이 양쪽에서 중앙으로 접근한다는 것이 다름.
    pivot_value = lst[left]
    l = left + 1
    r = right
    # Left Pivot, LR Pointers라고도 함.

    # 서로의 인덱스가 엇갈릴 때까지 반복.
    while l <= r:
        # 만약 l과 r이 서로 가리키지 말아햐 하는 값을 가리키면 둘을 교환. 
        if lst[l] > pivot_value and lst[r] < pivot_value:
            lst[l], lst[r] = lst[r], lst[l]
            l += 1
            r -= 1
        
        # l의 인덱스가 유효하면 1 증가.
        if lst[l] <= pivot_value:
            l += 1

        # r의 인덱스가 유효하면 1 증가.
        if lst[r] >= pivot_value:
            r -= 1
    
    # 최종적으로 r이 pivot_value보다 작거나 같은 값을 가리키게 된다.
    lst[r], lst[left] = lst[left], lst[r]
    return r
````

````python
# 각각의 파티션 방법을 사용한 quick_select 정의.
def quick_select_lomuto_rec(lst, k):
    return __quick_select_rec(lst, k, __partition_lomuto)

def quick_select_hoare_rec(lst, k):
    return __quick_select_rec(lst, k, __partition_hoare)
````

````python
# 반복적인 quick_select 함수의 원형.
def __quick_select_iter(lst, k, partition):
    # left와 right의 초기 설정.
    left = 0
    right = len(lst) - 1

    # k가 0 이하거나 len(lst)초과가 아닌 경우 알고리즘이 종료될 것이 보장되어 있으므로 while True를 사용할 수 있음.
    while True:
        # 매 반복마다 pos 갱신.
        pos = partition(lst, left, right)

        # 재귀적인 경우와 동일하나 재귀 호출을 통해 각 변수를 갱신하는 것이 아니라 직접 갱신.
        if pos + 1 == left + k:
            return lst[pos]
        elif pos + 1 > left + k:
            right = pos - 1
        else:
            k = k - (pos + 1 - left)
            left = pos + 1
````

````python
# 각각의 파티션 방법을 사용한 quick_select 정의.
def quick_select_lomuto_iter(lst, k):
    return __quick_select_iter(lst, k, __partition_lomuto)

def quick_select_hoare_iter(lst, k):
    return __quick_select_iter(lst, k, __partition_hoare)
````

##### # 로무토 파티션과 호어 파티션의 비교

\# 로무토 파티션
위에서 구현한 로무토 파티션은 Right Pivot, LL Pointers라고도 한다. 피봇의 기준을 오른쪽으로 정하고 두 포인터가 오른쪽에서부터 왼쪽으로 움직인다는 뜻이다.  
여기서 항상 다른 포인터보다 왼쪽에 있는 포인터를 l, 반대의 경우를 r이라고 하자.  

이때 l은 항상 pivot_value보다 작거나 같은 값을, r은 항상 pivot_value보다 큰 값을 가리키는 것이 이상적이다.  
따라서 이를 구현하기 위해 l은 출발선보다 한 칸 앞서게 배치하고, r을 먼저 출발시킨다.  

r이 선발대로 출발해 l이 가리킬 만한 것이 있는지 탐색한다.  
탐색하다 대상을 찾으면 l을 한 칸 앞으로 보낸 후 서로의 원소를 교환한다.  
이렇게 하면 l은 항상 pivot_value보다 작거나 같은 값을 가리키게 할 수 있다.  
이 과정에서 r은 pivot_value보다 큰 값은 무시하면서 지나가므로 자연스럽게 l과의 교환에서 l이 가리키고 있던 pivot_value보다 크거나 같은 값을 자신이 가리키게 할 수 있다.  

\# 호어 파티션
위에서 구현한 호어 파티션은 Left Pivot LR Pointers라고도 한다. 피봇의 기준을 왼쪽으로 정하고 두 포인터가 양쪽에서 중앙으로 움직인다는 뜻이다.  
여기서 항상 다른 포인터보다 왼쪽에 있는 포인터를 l, 반대의 경우를 r이라고 하자.  

이때 l은 항상 pivot_value보다 작거나 같은 값을, r은 항상 pivot_value보다 큰 값을 가리키는 것이 이상적이다.  
이번에는 l과 r을 동시에 출발시켜보자.  

l과 r이 각각 양쪽에서부터 탐색하면서 l과 r이 가리켜야 하는 값의 반대되는 원소를 가리키면 이 둘의 원소를 바꾼다.  
이를 l과 r이 엇갈릴 때까지 반복한다.  

l과 r이 엇갈리면 마지막 반복문에서 l과 r의 조건절을 통과하지 못했다는 의미가 되므로 l은 pivot_value보다 크거나 같은 값을 r은 pivot_value보다 작거나 같은 값을 가리키게 된다.  
이때 기준이 되었던 피봇이 왼쪽에 있었으므로 l이 아닌 r과 그 원소를 바꾼 다음 r을 피봇 인덱스로 반환한다.  

\# 두 알고리즘의 비교
로무토 파티션은 l과 r이 동일할 때에도 원소 교환을 수행한다. 따라서 이에 대한 추가적인 비용이 발생에 비교적 비싼 비용이 된다.  
반면에 호어 파티션은 필요한 경우가 아니라면 원소 교환을 수행하지 않기 때문에 이에 대한 추가적인 비용이 없다.  

즉, 원소 교환의 횟수가 비교적 적은 호어의 파티션 알고리즘이 더 효율적이다.  

````python
# 동전 문제 풀이.
def con_prob(coins):
    # 동전 12개르르 4개씩 3개의 그룹으로 나눔.
    group1 = [i for i in range(0, 4)]
    group2 = [i for i in range(4, 8)]
    group3 = [i for i in range(8, 12)]

    # 첫 두 그룹의 무게를 측정.
    w_comp = scale(coins, group1, group2)

    # 첫 두 그룹의 무게가 동일하면,
    if w_comp == 0:
        # 두 그룹은 정상적인 동전의 모임임.
        mark(coins, group1, 'normal')
        mark(coins, group2, 'normal')

        # 정상 그룹으로 분류한 그룹과 나머지 그룹을 비교하기 위해 각각의 그룹을 분리.
        # 이때 이후에 또 다른 비교를 위해 각 그룹에서 동적을 3개씩 각출.
        normal_group = [group1[i] for i in range(3)]
        check_group = [group3[i] for i in range(3)]

        # 정상 그룹과 비교 그룹의 무게를 측정.
        w_comp = scale(coins, check_group, normal_group)

        # 위의 그룹의 무게가 동일하면 처음에 제외했던 동전이 불량 동전이므로 이를 표시하고 종료.
        if w_comp == 0:
            # 무게가 더 적은지 큰지를 알아내기 위해 다시 한 번 무게 비교.
            coins[group3[3]]['check'] = 'less' if scale(coins, check_group[0], normal_group[3]) else 'more'
            return

        # 비교 그룹의 무게가 더 작은 경우,
        elif w_comp == -1:
            mark(coins, check_group, 'less_')
            w_comp = scale(coins, [check_group[0]], [check_group[1]])

            if w_comp == 0:
                mark(coins, check_group[2], 'less')
            elif w_comp == -1:
                mark(coins, check_group[2], 'less')
            elif w_comp == 1:
                mark(coins, check_group[2], 'less')
            return
        
        # 비교 그룹의 무게가 더 큰 경우,
        elif w_comp == 1:
            mark(coins, check_group, 'more_')
            return
````

``` c++
class Solution {
public:
    double findMedianSortedArrays(vector<int>& nums1, vector<int>& nums2) {
        // 두 벡터를 병합하여 중앙값을 탐색.
        vector<int> merged = mergeVectors(nums1, nums2);
        
        float median = merged[merged.size() / 2];

        // 벡터의 길이가 짝수이면 중앙값을 구하는 식이 달라지므로 이를 고려함.
        if (merged.size() % 2 == 0) {
            median += merged[merged.size() / 2 - 1];
            median /= 2;
        }
    
        return median;
    }
    
private:
    vector<int> mergeVectors(vector<int>& v1, vector<int>& v2) {
        // 두 벡터를 병합할 벡터 생성.
        vector<int> merged = vector<int>(v1.size() + v2.size());
        
        int i1 = 0;
        int i2 = 0;
        
        int index = 0;

        // 각 벡터를 가리키는 인덱스 중 하나가 완료될 때까지 반복.
        while (i1 < v1.size() && i2 < v2.size()) {
            // v1의 원소가 적당하면 v1의 원소 삽입.
            if (v1[i1] < v2[i2])
                merged[index++] = v1[i1++];

            // v2의 원소가 적당하면 v1의 원소 삽입.
            else
                merged[index++] = v2[i2++];
        }
        
        // 나머지 원소들 삽입.

        // v1의 원소가 모두 삽입되었으면 v2에서 나머지 원소 삽입.
        if (i1 == v1.size())
            while (i2 < v2.size())
                merged[index++] = v2[i2++];
                
        // v2의 원소가 모두 삽입되었으면 v1에서 나머지 원소 삽입.
        else
            while (i1 < v1.size())
                merged[index++] = v1[i1++];
        
        return merged;
    }
};
```
